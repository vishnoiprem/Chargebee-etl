```python
#Group by 10-20,20-40,40-60
#function can group 
def group_by_age(val):
  # print(val)
  tgt=''
  if  int(val)<=20:
    tgt='0-20'
    return tgt

  elif int(val)<=30:
      tgt='21-30'
      return tgt
  elif int(val)<=40:
      tgt='31-40'
      return tgt

  elif int(val)<=50:
      tgt='41-50'
      return tgt

  elif int(val)<=60:
      tgt='51-60'
      return tgt
  
  elif int(val)<=100:
      tgt='61-80'
      return tgt
  else:
    tgt='0-1'
    return tgt
```

```python
#new column
xls_df["age_group"]=xls_df['Age'].map(group_by_age)
```

```python
sql_run="""
select age_group,
       MALE,
       FEMALE,
       total
FROM (select age_group,
             COUNT(case when sex = 'M' THEN 'MALE' ELSE NULL END)   MALE,
             COUNT(case when sex = 'F' THEN 'FEMALE' ELSE NULL END) FEMALE,
             COUNT(*) as                                            total
      from xls_df
      group by age_group
     ) re
"""
```
